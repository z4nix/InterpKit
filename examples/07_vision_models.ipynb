# Vision Models

interpkit works on any HuggingFace model — not just language models.
This notebook demonstrates using interpkit with ResNet and ViT for image classification interpretability.

```python
import interpkit
```

## ResNet

```python
resnet = interpkit.load("microsoft/resnet-18")
resnet.inspect()
```

## Create a test image

```python
from PIL import Image
import tempfile
import os

# Create a simple test image (or use your own)
img = Image.new("RGB", (224, 224), color=(100, 150, 200))
img_path = os.path.join(tempfile.gettempdir(), "test_image.jpg")
img.save(img_path)
print(f"Test image saved to: {img_path}")
```

## Gradient attribution on images

For vision models, attribution produces a spatial heatmap showing which image regions matter most.

```python
resnet.attribute(img_path, save="resnet_attribution.png")
```

## Activations from conv layers

```python
act = resnet.activations(img_path, at="resnet.encoder.stages.0")
print(f"Stage 0 activations: {act.shape}")
```

## Ablation on vision models

How important is each stage of ResNet?

```python
for i in range(4):
    stage = f"resnet.encoder.stages.{i}"
    try:
        result = resnet.ablate(img_path, at=stage, method="zero")
        print(f"Stage {i}: effect = {result['effect']:.4f}")
    except Exception as e:
        print(f"Stage {i}: {e}")
```
