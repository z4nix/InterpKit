# Attention Patterns

Explore attention heads across layers — find which heads attend to which tokens,
export static heatmaps and interactive HTML visualizations.

```python
import mechkit

model = mechkit.load("gpt2")
```

## All layers — summary table

Running `attention()` with no filters shows the top attention pairs and entropy for every head.

```python
results = model.attention("The Eiffel Tower is located in the city of")
```

## Single layer

Filter to a specific layer to reduce noise.

```python
results = model.attention(
    "The Eiffel Tower is located in the city of",
    layer=5,
)
```

## Single head — heatmap

Zoom into a single head and export a matplotlib heatmap.

```python
results = model.attention(
    "The Eiffel Tower is located in the city of",
    layer=5,
    head=0,
    save="attention_L5H0.png",
)
```

## Interactive HTML

Generate a self-contained HTML file with hover tooltips, layer dropdown, and per-head grids.

```python
results = model.attention(
    "The Eiffel Tower is located in the city of",
    html="attention_interactive.html",
)
```

## Working with the raw weights

Each result entry contains the full attention weight matrix as a tensor.

```python
import torch

# Get attention for layer 0
layer0_heads = [r for r in results if r["layer"] == 0]

for head_data in layer0_heads[:3]:
    w = head_data["weights"]
    if isinstance(w, torch.Tensor):
        print(f"Layer {head_data['layer']}, Head {head_data['head']}: "
              f"shape={w.shape}, entropy={head_data['entropy']:.3f}")
```
